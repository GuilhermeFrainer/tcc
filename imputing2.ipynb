O objetivo do código aqui é imputar valores para a série de consumo até jan/2003

```python
import pandas as pd
import numpy as np
import utils
from sktime.transformations.series.impute import Imputer
```

```python
df = pd.read_csv("dados/dados_crus2003.csv", sep=";", decimal=",")
df = df.drop(columns=df.columns.difference(["month", "energia_total", "energia_residencial", "energia_industrial", "energia_comercial", "energia_outros"]))
df = utils.index_to_period(df)
df
```

```python
for col in df:
    transformer = Imputer()
    df[col] = transformer.fit_transform(df[col])

df.head(15)
```

```python
df.to_csv("dados/energia_cru_2003.csv", sep=";", decimal=",")
```
