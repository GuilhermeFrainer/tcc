O objetivo do código aqui é fazer cópias dos dados em `dados_transformados2003.csv` com lags

```python
import pandas as pd
import utils
```

```python
df = pd.read_csv("dados/dados_transformados2003.csv", sep=";", decimal=",")
df = utils.index_to_period(df)
shifted_df = df.copy()
for col in df:
    for i in range(10):
        shifted_df[f'{col}_lag{i + 1}'] = df[col].shift(i + 1)
shifted_df.head()
```

```python
shifted_df = shifted_df.dropna()
shifted_df
```

```python
shifted_df.to_csv("dados/lags_transformados.csv", sep=";", decimal=",")
```
